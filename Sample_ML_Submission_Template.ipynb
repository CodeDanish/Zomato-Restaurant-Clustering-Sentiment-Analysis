## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# Create copies to preserve original data
meta_clean = res_meta.copy()
reviews_clean = res_reviews.copy()
````

````python
#Information about Restaurant MetaData before cleaning
meta_clean.info()
````

````python
# =====================================================================
# 1. CLEANING RESTAURANT METADATA (`meta_clean`)
# =====================================================================

# Here we don't have any duplicate values as found during profiling

# Fix 'Cost' column: Remove commas and convert to numeric float
meta_clean["Cost"] = (
    meta_clean["Cost"]
    .str.replace(",", "")
    .str.strip())

meta_clean["Cost"] = pd.to_numeric(meta_clean["Cost"], errors="coerce")

# Fill missing 'Collections' with 'Standard' or 'None' tag
meta_clean["Collections"].fillna("No Collection Tag", inplace=True)

# Fill missing 'Timings' with a standard placeholder
meta_clean["Timings"].fillna("Timings Not Available", inplace=True)

# Clean up temporary columns or unnecessary spaces from joined IDs
meta_clean["Name"] = meta_clean["Name"].str.strip()
````

````python
# Info View of the "meta_clean" Dataset after Cleaning
meta_clean.info()
````

````python
# Information about Reviews Dataset before cleaning
reviews_clean.info()
````

````python
# =====================================================================
# 2. CLEANING REVIEWS DATASET (`reviews_clean`)
# =====================================================================

# Drop duplicate rows found during profiling (36)
reviews_clean.drop_duplicates(inplace=True)

# Drop rows where critical modeling fields ('Review' text or 'Rating') are missing
reviews_clean.dropna(subset=["Review", "Rating"], inplace=True)

# Convert 'Rating' to numeric float (handles string scores like '4.5' or mixed types)
reviews_clean["Rating"] = pd.to_numeric(
    reviews_clean["Rating"], errors="coerce"
)
# Drop rows where rating couldn't be converted (e.g., text corruptions like 'Like')
reviews_clean.dropna(subset=["Rating"], inplace=True)

# Convert 'Time' column into a proper datetime format for time-series analysis
reviews_clean["Time"] = pd.to_datetime(
    reviews_clean["Time"], errors="coerce"
)
# Drop null values from the 'Time' column
reviews_clean.dropna(subset=["Time"], inplace=True)
````

````python
# Info view of the Reviews Dataset after Data Cleaning
reviews_clean.info()
````

````python
# =====================================================================
# 3. FEATURE ENGINEERING FROM TEXT FIELDS
# =====================================================================

# Extract 'Review_Count' and 'Followers_Count' from the unstructured 'Metadata' column
# Format example: "3 Reviews , 2 Followers"
reviews_clean["Review Count"] = (
    reviews_clean["Metadata"]
    .str.extract(r"(\d+)\s+Review")
    .fillna(0)
    .astype(int)
)
reviews_clean["Followers Count"] = (
    reviews_clean["Metadata"]
    .str.extract(r"(\d+)\s+Follower")
    .fillna(0)
    .astype(int)
)

# Clean up temporary columns or unnecessary spaces from joined IDs
reviews_clean["Restaurant"] = reviews_clean["Restaurant"].str.strip()
````

````python
# View of 'Review_Count','Followers_Count' and 'Restaurant' columns
print(reviews_clean["Review Count"].values[0:10])
````

````python
print(reviews_clean["Followers Count"].values[0:10])
````

````python
print(np.unique(reviews_clean["Restaurant"].values[0:1000]))
````

````python
# =====================================================================
# 4. MERGING THE DATASETS FOR AN INTEGRATED ANALYSIS PIPELINE
# =====================================================================

# Merge datasets using a left join to align restaurant metadata with review dynamics
zomato_final = pd.merge(
    meta_clean,
    reviews_clean,
    left_on="Name",
    right_on="Restaurant",
    how="left",
)

# Dropping Name and Metadata columns from the Merged dataset
zomato_final.drop(columns=["Restaurant","Metadata"],inplace=True)

# Renaming the "Cost" and "Time" columns
zomato_final.rename(columns={"Cost":"Cost Per Person", "Time":"Review Time"},inplace=True)

# Verify results
print("--- Wrangling Pipeline Complete ---")
print()
print(f"Final Integrated Dataset Shape: {zomato_final.shape}")
print()
print("View of the Final Integrated Dataset :")
zomato_final.head()
````

````python
# Information about the Final Integrated Dataset
zomato_final.info()
````

````python
# Dropping the Null values
zomato_final.dropna(inplace=True)

zomato_final.info()
````

* Converted raw text timestamps (`Time`) into standardized pandas `datetime` format.

3. **Feature Engineering (Regex Text Extraction):** * Deconstructed the cluttered `Metadata` text column using regular expressions (`\d+`) to isolate and build two independent numeric metrics: `Review Count` and `Followers Count`.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

To statistically evaluate **Hypothetical Statement - 1 (The Premium Value Paradox)**, we need to mathematically test whether restaurants in Zomato's curated marketing "Collections" actually maintain a higher baseline of customer ratings compared to standard, non-curated listings.

Here is the structured formulation of our research hypotheses:

### **Null Hypothesis ($H_0$):**

* **Statement:** There is no significant difference between the average customer ratings of curated track restaurants and standard (non-curated) restaurants, or standard restaurants actually have equal/higher average ratings.
* **Mathematical Representation:**
$$H_0: \mu_{\text{Curated}} \le \mu_{\text{Standard}}$$



*(where $\mu$ represents the population mean of customer ratings)*

### **Alternative Hypothesis ($H_1$ or $H_a$):**

* **Statement:** Restaurants featured in Zomato's curated marketing "Collections" have a statistically significant, higher average customer rating compared to standard, non-curated listings.
* **Mathematical Representation:**
$$H_1: \mu_{\text{Curated}} > \mu_{\text{Standard}}$$



*(This sets up a **One-Tailed (Right-Tailed) Independent Two-Sample T-Test** or a non-parametric Mann-Whitney U Test, depending on the distribution's normality).*

#### 2. Perform an appropriate statistical test.

Here we perform an independent Welch's Two-Sample t-test (which accounts for unequal sample sizes and variances) to verify your hypothesis and calculate the exact $p$-value.

````python
# Perform Statistical Test to obtain P-Value
import numpy as np
import pandas as pd
from scipy import stats

# 1. Establish the Curated vs Standard grouping flag in your dataset
# (Assumes your wrangled data framework or original data handles the collection strings)
zomato_final["Is_Curated"] = zomato_final["Collections"].apply(
    lambda x: False
    if (x == "No Collection Tag" or pd.isna(x) or x == "None")
    else True
)

# 2. Extract the rating distributions for both mutually exclusive groups
curated_ratings = zomato_final[zomato_final["Is_Curated"] == True]["Rating"]
standard_ratings = zomato_final[zomato_final["Is_Curated"] == False]["Rating"]

# 3. Calculate descriptive metrics for baseline verification
print("--- Sample Profiles ---")
print(
    f"Curated Track Group  -> Count: {len(curated_ratings)}, Mean Rating: {curated_ratings.mean():.4f}"
)
print(
    f"Standard Group  -> Count: {len(standard_ratings)}, Mean Rating: {standard_ratings.mean():.4f}"
)
print()

# 4. Perform Independent Welch's t-test (equal_var=False handles variance mismatch)
t_stat, p_val_two_tailed = stats.ttest_ind(
    curated_ratings, standard_ratings, equal_var=False
)

# 5. Convert to a One-Tailed (Right-Tailed) test since H1 states Curated > Standard
if t_stat > 0:
    p_value_one_tailed = p_val_two_tailed / 2
else:
    p_value_one_tailed = 1 - (p_val_two_tailed / 2)

# 6. Output the Statistical Results
print("--- Statistical Test Metrics ---")
print(f"Calculated T-Statistic : {t_stat:.4f}")
print(f"Calculated P-Value    : {p_value_one_tailed:.4e}")
````

##### Which statistical test have you done to obtain P-Value?

To calculate the p-value, an Independent Two-Sample Welch’s t-test (one-tailed, right-sided configuration) was performed.

##### Why did you choose the specific statistical test?

The choice of Welch's t-test was based on the following structural characteristics of the dataset:
*  **Continuous Dependent Variable:** The metric being evaluated (Rating) is an interval/ratio numeric scale, which matches the target input requirements for a t-test.

*  **Categorical Independent Groups:** The data naturally splits into two mutually exclusive, independent comparison groups based on platform features: **Group A** (Restaurants with a Collections tag) and **Group B** (Restaurants without a Collections tag).

*  **Unequal Sample Variances & Sizes (Welch's Adaptation):** A standard Student’s t-test assumes that both comparison groups share identical variances. However, your data segments are inherently unbalanced due to varying organic review volumes. Welch's t-test does not assume equal variances, protecting the resulting p-value from inflation or structural distortion.

*  **Directional Research Intent:** Because the hypothesis explicitly states that curated track options score higher ($\mu_{\text{Curated}} > \mu_{\text{Standard}}$) rather than just different ($\mu_{\text{Curated}} \neq \mu_{\text{Standard}}$), a **one-tailed (right-tailed)** variant was used to maximize statistical power.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

To statistically evaluate **Hypothetical Statement - 2 (Visual Media Domination by the "Critique Elite")**, we want to test whether a reviewer's platform status—specifically their social equity (`Followers Count`)—has a significant positive relationship with the volume of visual media (`Pictures`) they upload.

Here is the structured formulation of your research hypotheses:

### **Null Hypothesis ($H_0$):**

* **Statement:** There is no linear relationship between a reviewer’s follower count and the number of pictures they upload to a restaurant review, or the relationship is negative.
* **Mathematical Representation:**
$$H_0: \rho \le 0$$



*(where $\rho$ represents the population Pearson or Spearman rank correlation coefficient between `Followers Count` and `Pictures`)*

### **Alternative Hypothesis ($H_1$ or $H_a$):**

* **Statement:** There is a statistically significant, positive relationship between a reviewer’s follower count and the number of pictures they upload.
* **Mathematical Representation:**
$$H_1: \rho > 0$$



*(This sets up a directional, **one-tailed correlation significance test**).*

#### 2. Perform an appropriate statistical test.

Because user engagement metrics like Followers Count and Pictures uploaded are heavily right-skewed, a non-parametric Spearman Rank Correlation Test is chosen to safely evaluate this relationship without being distorted by extreme outliers.

````python
# Perform Statistical Test to obtain P-Value
import numpy as np
import pandas as pd
from scipy import stats

# 1. Clean and handle potential missing values in target columns
test_data = zomato_final[["Followers Count", "Pictures"]].dropna()

# 2. Execute the Spearman Rank Correlation Test
# (Handles skewed data metrics much better than standard Pearson)
rho, p_val_two_tailed = stats.spearmanr(
    test_data["Followers Count"], test_data["Pictures"]
)

# 3. Adapt to a One-Tailed configuration since H1 tests explicitly for a positive direction (rho > 0)
if rho > 0:
    p_value_one_tailed = p_val_two_tailed / 2
else:
    p_value_one_tailed = 1 - (p_val_two_tailed / 2)

# 4. Display the calculation output
print("--- Correlation Statistical Metrics ---")
print(f"Spearman Rho (Correlation Coeff): {rho:.4f}")
print(f"Calculated P-Value               : {p_value_one_tailed:.4e}")
````

##### Which statistical test have you done to obtain P-Value?

To calculate the p-value, a One-Tailed Spearman’s Rank Correlation Test was performed.

##### Why did you choose the specific statistical test?

The choice of the Spearman rank correlation coefficient over other metrics (like Pearson's $r$) was driven by the following characteristics of your dataset:

Instead, they follow a power-law distribution where the vast majority of users have near-zero metrics, while a tiny handful of elite critics have massive, extreme values. Standard Pearson correlation breaks down with non-normal data and is highly distorted by outliers.

*  **Monotonic vs. Linear Relationships:** Your hypothesis does not require a perfectly straight line relationship (e.g., 100 more followers exactly equals 50 more pictures). It tests for a **monotonic relationship** —meaning as one variable increases, the other variable consistently tends to increase as well, regardless of the rate. Spearman captures this perfectly by converting raw values into ordinal ranks before calculating the correlation.

*  **Ordinal Rank Evaluation:** By ranking the data, the test evaluates if high-ranked influencers (Rank 1, 2, 3 in followers) naturally align with high-ranked visual contributors (Rank 1, 2, 3 in photo uploads), successfully isolating the "Critique Elite" behavioral footprint.

* **Directional Research Intent:** Because the alternative hypothesis specifically expects a positive direction ($\rho > 0$), a **one-tailed** focus was applied to increase the statistical power of the calculation.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

To statistically evaluate **Hypothetical Statement - 3 (The Weekend Operational Friction)**, we need to test whether the relationship between a restaurant's pricing (`Cost Per Person`) and its customer satisfaction (`Rating`) structurally changes or degrades during high-stress weekend operations compared to calm weekdays.

Here is the structured formulation of your research hypotheses:

### **Null Hypothesis ($H_0$):**

* **Statement:** The impact of a restaurant's cost per person on its customer rating is identical on weekends and weekdays, or weekends do not show a lower rating performance across pricing tiers compared to weekdays.
* **Mathematical Representation:**
$$H_0: \beta_{\text{Weekend}} \ge \beta_{\text{Weekday}}$$



*(where $\beta$ represents the regression slope or interaction coefficient mapping how `Cost Per Person` predicts `Rating` across both day types)*

### **Alternative Hypothesis ($H_1$ or $H_a$):**

* **Statement:** High-cost restaurants experience a statistically significant downward shift or drop in customer ratings during weekend operations compared to weekday operations.
* **Mathematical Representation:**
$$H_1: \beta_{\text{Weekend}} < \beta_{\text{Weekday}}$$



*(This sets up a directional, **one-tailed interaction analysis** or a two-way ANOVA/Regression framework to isolate the compounding negative effects of cost and weekends on overall ratings).*

#### 2. Perform an appropriate statistical test.

To evaluate whether the relationship between price and rating changes significantly on weekends versus weekdays, we can use an Ordinary Least Squares (OLS) Linear Regression model with an interaction term.

The interaction term (Cost * Is_Weekend) directly tests if the slope of the cost-to-rating relationship changes significantly depending on the day of the week.

````python
# Perform Statistical Test to obtain P-Value
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 1. Feature Engineering: Create a binary flag for Weekends (Friday=4, Saturday=5, Sunday=6)
# (Assumes your wrangled dataframe is named 'zomato_final')
zomato_reg = zomato_final.copy()
zomato_reg["Is_Weekend"] = (
    zomato_reg["Review Time"].dt.dayofweek.isin([4, 5, 6]).astype(int)
)

# 2. Rename columns to remove spaces for the patsy formula notation
zomato_reg = zomato_reg.rename(
    columns={"Cost Per Person": "Cost_Per_Person", "Review Time": "Review_Time"}
)

# 3. Fit an OLS model with an interaction term: Rating ~ Cost + Is_Weekend + (Cost * Is_Weekend)
# The ':' operator creates the interaction variable automatically
model = smf.ols(
    formula="Rating ~ Cost_Per_Person * Is_Weekend", data=zomato_reg
).fit()

# 4. Print the comprehensive regression metrics summary
print(model.summary())

# 5. Isolate the specific interaction P-Value
interaction_p_value = model.pvalues["Cost_Per_Person:Is_Weekend"]
interaction_coef = model.params["Cost_Per_Person:Is_Weekend"]

print("\n" + "=" * 50)
print(f"Interaction Coefficient (Slope Shift): {interaction_coef:.6f}")
print(f"Extracted Interaction P-Value       : {interaction_p_value:.4e}")
print("=" * 50)
````

##### Which statistical test have you done to obtain P-Value?

To calculate the p-value, an Ordinary Least Squares (OLS) Linear Regression Analysis with a Multiplicative Interaction Term was performed.

##### Why did you choose the specific statistical test?

The choice of an interactive linear regression model over simpler comparative tests (like a basic t-test or ANOVA) was driven by the following analytical requirements:
*  **Testing an Interaction Effect (Moderation Analysis):** The hypothesis does not simply state that weekend ratings are lower than weekday ratings. It states that the relationship between price and rating shifts based on the day of the week. In statistics, this means the day of the week acts as a **moderator variable**. An interaction term (Cost * Is_Weekend) is the mathematically correct way to measure if a third variable alters the slope of the primary relationship.

*  **Mixed Variable Types:** Your features consist of a continuous numeric independent variable (Cost Per Person), a binary categorical independent moderator (Is_Weekend: Yes/No), and a continuous numeric dependent output (Rating). Multiple linear regression naturally blends continuous and discrete parameters within a unified model.

*  **Quantifying the Exact Slope Deviation:** Rather than just telling you if a difference exists, this test provides an exact numerical coefficient ($\beta$) for the interaction term. This coefficient tells Zomato exactly how much a restaurant's rating drops for every additional dollar/INR spent when shifting operations from a quiet weekday to a high-volume weekend night.

*  **Directional Target Focus:** By observing if the specific interaction coefficient is negative ($\beta < 0$) alongside a significant p-value, we can directly confirm or deny the exact directional claim of **Weekend Operational Friction**.

````python
# ==========================================
# 1. PROCESSING THE METADATA DATAFRAME
# ==========================================

# A. Handle 'Collections' (Categorical)
# Missing values imply the restaurant wasn't picked for a curated track.
# Instead of dropping, we impute a meaningful structural placeholder.
res_meta["Collections"] = res_meta["Collections"].fillna("No Collection Tag")

# B. Handle 'Timings' (Categorical)
# Missing operational hours can be labeled safely as 'Unknown' to preserve rows.
res_meta["Timings"] = res_meta["Timings"].fillna("Unknown Timings")

# C. Handle 'Cuisines' (Categorical)
# Cuisines are critical for analytics. If missing, we fall back on the most frequent
# cuisine type (Mode) across the dataset, or a placeholder if preferred.
cuisine_mode = res_meta["Cuisines"].mode()[0]
res_meta["Cuisines"] = res_meta["Cuisines"].fillna(cuisine_mode)

# D. Handle 'Cost Per Person' (Numerical)
# Critical Step: Replacing cost with a global mean/median causes extreme distortion.
# Instead, we impute using a 'Grouped Median' based on the primary cuisine type.


def clean_and_impute_cost(df):
    # Ensure cost is numeric (handling string formats like '1,200')
    if df["Cost"].dtype == "O":
        df["Cost"] = (
            df["Cost"]
            .str.replace(",", "")
            .str.extract(r"(\d+)")
        )
        df["Cost"] = pd.to_numeric(
            df["Cost"], errors="coerce"
        )

    # Calculate median cost per primary cuisine cluster
    df["Primary_Cuisine"] = df["Cuisines"].astype(str).str.split(",").str[0]
    cuisine_medians = df.groupby("Primary_Cuisine")["Cost"].transform(
        "median"
    )

    # Fallback global median for any unique cuisines without a calculated median
    global_median = df["Cost"].median()

    # Apply the localized median imputation
    df["Cost"] = df["Cost"].fillna(
        cuisine_medians.fillna(global_median)
    )
    df.drop(columns=["Primary_Cuisine"], inplace=True)
    return df


res_meta = clean_and_impute_cost(res_meta)
````

* **Why it was used:** Imputing or guessing a target variable injects artificial bias into your data. Since a missing target cannot be safely reconstructed, dropping these specific rows ensures your statistical calculations and machine learning models remain clean and reliable.

````python
# ==========================================
# 1. VISUAL INSPECTION (DIAGNOSTIC CRITIQUE)
# ==========================================
def plot_outlier_diagnostics(df, column, title_name):
    """Generates a side-by-side Boxplot and Distribution plot to isolate outliers."""
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    sns.boxplot(x=df[column], ax=axes[0], color="#CB202D")
    axes[0].set_title(f"{title_name} - Boxplot")

    sns.histplot(df[column], kde=True, ax=axes[1], color="#1F1F1F")
    axes[1].set_title(f"{title_name} - Distribution")

    plt.tight_layout()
    plt.show()


# Run initial diagnostics on pricing before treatment
print("=== Diagnostic: Pre-Treatment Structural Skew ===")
plot_outlier_diagnostics(zomato_final, "Cost Per Person", "Raw Cost Per Person")
````

````python
# ==========================================
# 2. TREATMENT TECHNIQUE A: IQR WINSORIZATION (CAPPING)
# ==========================================
def treat_outliers_iqr_capping(df, column):
    """Caps extreme outliers at the 1.

    5*IQR boundaries to neutralize their distorting influence.
    """
    df_treated = df.copy()

    # Calculate Interquartile Range (IQR) boundaries
    q1 = df_treated[column].quantile(0.25)
    q3 = df_treated[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Logical Adjustment: Ensure the lower bound never drops below 0 for structural realism
    lower_bound = max(0, lower_bound)

    print(f"--- IQR Bounds for '{column}' ---")
    print(
        f"Lower Baseline Bound: {lower_bound:.2f} | Upper Baseline Bound: {upper_bound:.2f}"
    )

    # Apply structural capping (Winsorization) using numpy clip
    df_treated[column] = np.clip(df_treated[column], lower_bound, upper_bound)

    return df_treated


# Apply IQR Capping to restaurant costs
zomato_final = treat_outliers_iqr_capping(zomato_final, "Cost Per Person")
````

````python
# ==========================================
# 3. TREATMENT TECHNIQUE B: LOG TRANSFORM (ENGAGEMENT METRICS)
# ==========================================
# User follower metrics follow a severe power-law distribution.
# Dropping them deletes your power-users, and capping destroys the variance.
# We apply a Log1p transformation ($log(x+1)$) to stabilize the scale.

zomato_final["Log_Followers"] = np.log1p(zomato_final["Followers Count"])
````

### 1. IQR Winsorization (Capping)

* **Technique:** Extreme values past the upper fence ($Q3 + 1.5 \times IQR$) are capped and pulled back to that exact maximum boundary line, rather than being deleted.
* **Applied To:** `Cost Per Person`
* **Why it was used:** Ultra-luxury fine dining restaurants create massive outliers on price distributions. If you leave them raw, they warp your model's linear relationships; if you delete them, you lose valuable restaurant entries. Capping neutralizes their distorting mathematical leverage while keeping the actual data rows fully intact.


### 2. Log1p Transformation ($\log(x + 1)$)

* **Technique:** A logarithmic scale transformation that compresses a highly stretched, right-skewed tail into a stabilized, near-normal bell curve.
* **Applied To:** User engagement metrics like `Followers` and `Pictures`.
* **Why it was used:** Platform social data follows a power-law distribution—the vast majority of users have 0 to 5 followers, while a tiny handful of "elite critics" have thousands. Standard capping fails here because it erases the true profile of your power users. A log transformation shrinks the vast scale difference between standard users and influencers, making the variance manageable for statistical and machine learning algorithms without discarding your platform's most important critics.

### 3. Categorical Encoding

For platform analytics, we use a hybrid encoding strategy to match the mathematical nature of different columns: One-Hot Encoding for low-cardinality nominal columns, Ordinal Encoding if a column has an inherent tier-based order, and Target/Mean Encoding for high-cardinality columns like Cuisines or City to prevent the dataset from blowing up into thousands of sparse columns.

````python
# Encode your categorical columns
````

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Assuming your clean, pre-processed dataframe is named 'zomato_final'
zomato_encoded = zomato_final.copy()

# ==========================================================
# TECHNIQUE 1: ONE-HOT ENCODING (For Low-Cardinality Categoricals)
# ==========================================================
# Best for columns with few unique values and no intrinsic ordering.
# 'Is_Curated' or 'Has_Table_Booking' fit perfectly here.

nominal_cols = ["Is_Curated"]  # Add other binary/low-category flags here

# Initialize OneHotEncoder (drop='first' prevents multicollinearity/dummy variable trap)
oh_encoder = OneHotEncoder(drop="first", sparse_output=False, dtype=int)

# Fit and transform
oh_encoded_vars = oh_encoder.fit_transform(zomato_encoded[nominal_cols])

# Convert back to a DataFrame with clear column naming
oh_encoded_df = pd.DataFrame(
    oh_encoded_vars,
    columns=oh_encoder.get_feature_names_out(nominal_cols),
    index=zomato_encoded.index,
)

# Merge back and drop the original raw text column
zomato_encoded = pd.concat([zomato_encoded, oh_encoded_df], axis=1).drop(
    columns=nominal_cols
)
````

````python
# ==========================================================
# TECHNIQUE 2: ORDINAL ENCODING (For Ordered Categoricals)
# ==========================================================
# If your dataset contains explicit tier systems like Price Range tiers (e.g., Cheap, Medium, Luxury)
# We manually define the hierarchy matrix to ensure numbers map correctly to value.

if "Price_Tier" in zomato_encoded.columns:
    tier_order = [["Cheap", "Medium", "Luxury"]]
    ord_encoder = OrdinalEncoder(categories=tier_order)

    zomato_encoded["Price_Tier_Encoded"] = ord_encoder.fit_transform(
        zomato_encoded[["Price_Tier"]]
    )
    zomato_encoded.drop(columns=["Price_Tier"], inplace=True)
````

````python
# ==========================================================
# TECHNIQUE 3: TARGET ENCODING (For High-Cardinality Categoricals)
# ==========================================================
# Columns like 'Cuisines' or 'Location' have dozens or hundreds of unique text variations.
# One-hot encoding them creates thousands of empty columns.
# Instead, we map each category to the global mean of our target metric ('Rating').

high_cardinality_cols = ["Cuisines"]

for col in high_cardinality_cols:
    # Calculate the mean target rating for every distinct category group
    category_means = zomato_encoded.groupby(col)["Rating"].mean()

    # Map the mean back onto the dataset to replace the raw text string
    zomato_encoded[f"{col}_Target_Encoded"] = zomato_encoded[col].map(
        category_means
    )

    # Fallback/Imputation: If a category has only 1 entry, fill with the global baseline mean rating
    global_rating_mean = zomato_encoded["Rating"].mean()
    zomato_encoded[f"{col}_Target_Encoded"] = zomato_encoded[
        f"{col}_Target_Encoded"
    ].fillna(global_rating_mean)

    # Remove the unreadable text column
    zomato_encoded.drop(columns=[col], inplace=True)


# ==========================================================
# VERIFICATION PIPELINE
# ==========================================================
print("=== Matrix Check: Encoded Features Output ===")
print(zomato_encoded.info())
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

### 1. One-Hot Encoding (with `drop='first'`)

* **Technique:** Converts categorical strings into individual binary columns ($1$ or $0$). The first category is dropped to prevent perfect multicollinearity.
* **Applied To:** Low-cardinality nominal variables like `Is_Curated` or operational flags (e.g., `Has_Table_Booking`).
* **Why it was used:** These categories have no mathematical order or rank. One-hot encoding allows algorithms to interpret them as distinct binary states without falsely assuming one category is "greater" than another. Dropping the first column ensures your regression models don't crash from overlapping multi-variable dependencies.


### 2. Target Encoding (Mean Encoding)

* **Technique:** Replaces each text category with the average value of the target variable (`Rating`) for that specific group.
* **Applied To:** High-cardinality nominal columns like `Cuisines` or `Location`.
* **Why it was used:** A column like `Cuisines` can have hundreds of unique combinations. If you used One-Hot Encoding here, your dataset would blow up into hundreds of sparse, empty columns full of zeros (**The Curse of Dimensionality**), destroying your model's computational speed. Target encoding compresses all that complex text data into a single, highly predictive column that represents historical quality benchmarks.


### 3. Ordinal Encoding

* **Technique:** Maps structured text categories to an ordered sequence of integers (e.g., `Cheap` $\rightarrow 0$, `Medium` $\rightarrow 1$, `Luxury` $\rightarrow 2$).
* **Applied To:** Explicit tier systems like price brackets (`Price_Tier`).
* **Why it was used:** Unlike cuisines, pricing tiers have an inherent mathematical hierarchy. Luxury is strictly higher than cheap. Ordinal encoding preserves this structured progression, allowing statistical and machine learning models to capture the natural scale and sequence of the tiers.

### Why Lemmatization Was Chosen Over Stemming

In NLP, you generally choose between two normalization techniques: **Stemming** and **Lemmatization**. Lemmatization is the superior choice for a Zomato reviews dataset for the following reasons:

* **Preserves Semantic Meaning:** Stemming uses crude, rule-based heuristics to chop off the ends of words. For example, a stemmer might cut *"studying"* and *"studies"* down to *"studi"*, which is not a real word. Lemmatization uses a full vocabulary dictionary and morphological analysis to return the actual dictionary base form (the *lemma*).

* **Handles Irregular Verbs and Multi-inflections:** A stemmer cannot link completely different looking words that share a root meaning. For instance, a stemmer fails to connect *"good"*, *"better"*, and *"best"*. A lemmatizer correctly normalizes them all back to **"good"**. For restaurant reviews where users use varied adjectives (e.g., *"ate"*, *"eating"*, *"eats"* $\rightarrow$ **"eat"**), this is crucial.

* **Improves Downstream Analytics:** Because lemmatization outputs real, grammatically correct words, your final text clustering topics, word clouds, and sentiment analysis features remain human-readable and interpretably precise.

### Why TF-IDF Was Chosen Over Other Techniques

When converting text to numbers, the two most common baseline choices are **Bag of Words (CountVectorizer)** and **TF-IDF**. TF-IDF is significantly better suited for a Zomato reviews dataset for several reasons:

#### 1. It Penalizes Globally Ubiquitous Words

A standard Bag of Words approach simply counts word occurrences. In restaurant reviews, words like *"food"*, *"place"*, *"restaurant"*, or *"ordered"* appear in almost every single entry. CountVectorizer gives these words massive numerical weight just because they are frequent, even though they tell your model absolutely nothing about whether the review is positive or negative.

TF-IDF evaluates a word's importance using a balancing formula:

$$\text{TF-IDF} = \text{Term Frequency (TF)} \times \text{Inverse Document Frequency (IDF)}$$

If a word appears across almost every review in your dataset, its IDF score drops close to zero, effectively neutralizing its weight.

#### 2. It Highlights Highly Descriptive Sentiment Signals

Because of how the math works, TF-IDF boosts the numerical weight of words that appear frequently within a *specific* review but are rare across the *rest* of the dataset. Critical sentiment drivers like *"overpriced"*, *"undercooked"*, *"exquisite"*, or *"hidden-gem"* receive much higher mathematical importance. This directly helps your downstream machine learning models pick up on the actual emotional core of the text.

#### 3. It Naturally Handles Document Length Variations

People write reviews of wildly different lengths—some leave a quick sentence, while others write a multi-paragraph essay. CountVectorizer inherently biases models toward longer reviews because they naturally contain more total words. TF-IDF includes an internal normalization step that scales the values down, ensuring a long, rambling review doesn't mathematically overpower a short, punchy, high-impact review.

###Why Data Transformation is Required
Our engineered feature pool contains variables with completely different scales, units, and distributions:

* Review_To_Follower_Ratio scales as a tiny fraction between 0 and 1.
* Clean_Review_Word_Count and Followers Count are heavily right-skewed (a few users/reviews have massive counts, while the vast majority have very small counts).

###The Transformations Used & Why
We apply a two-step transformation strategy:

####1. Log Transformation (For Heavily Skewed Features)
**Applied to:** Followers Count, Review Count, and Clean_Review_Word_Count.

**Why:** These features follow a power-law distribution. Applying a Log transformation $\log(x + 1)$ compresses extreme outliers and pulls the long tail inward, converting a highly skewed distribution into a more symmetric, normal-like distribution. This makes it much easier for linear and tree-based models to find patterns.

####2. Robust Scaling (For Scale Standardization)
**Applied to:** All numerical inputs.

**Why:** Standardizers like MinMaxScaler compress normal data if extreme outliers exist. StandardScaler shifts data based on the mean, which outliers can easily distort. RobustScaler uses the median and the Interquartile Range (IQR), making it completely immune to extreme restaurant outlier spikes while bringing all features onto a uniform, comparable scale.

````python
# Transform Your data
from sklearn.preprocessing import RobustScaler

# Create a copy of our selected feature dataset
df_transformed = X_filtered.copy()

print("=== Phase 3: Data Transformation Pipeline ===")

# ==========================================================
# 1. LOG TRANSFORMATION (Handling Extreme Right-Skewness)
# ==========================================================
skewed_features = ["Followers Count", "Review Count", "Clean_Review_Word_Count"]

# Apply log1p (log(x+1)) to safely handle any zero counts without crashing
for col in skewed_features:
    if col in df_transformed.columns:
        df_transformed[col] = np.log1p(df_transformed[col])

print(" Log transformation completed for skewed features.")
````

````python
# ==========================================================
# 2. ROBUST SCALING (Outlier-Resilient Scale Standardization)
# ==========================================================
# Isolate numerical features that need uniform scaling
scale_features = [
    col
    for col in df_transformed.columns
    if col not in ["Is_Dinner_Rush", "Is_Lunch_Rush", "Is_Curated_True"]
]

scaler = RobustScaler()
df_transformed[scale_features] = scaler.fit_transform(
    df_transformed[scale_features]
)

print(" Robust Scaling scale alignment completed.")

# ==========================================================
# VERIFICATION AUDIT
# ==========================================================
print("\n--- Transformed Data Sample (Uniformly Scaled & Normalized) ---")
print(df_transformed[scale_features].head(3))
````

### Why Robust Scaling Was Chosen Over Other Methods

When prepping data for machine learning, the three main choices are **MinMax Scaling**, **Standard Scaling (Z-score)**, and **Robust Scaling**. Robust Scaling is the superior choice for a restaurant platform dataset like Zomato for the following reasons:

#### 1. It is Immune to Extreme Outliers

Zomato data contains severe outliers—for example, a few hyper-premium restaurants have massive costs per person, or a handful of viral food items have an overwhelming number of picture uploads compared to the average local diner.

* **MinMaxScaler** would compress the vast majority of normal restaurants into a tiny, squashed fraction near zero just to accommodate those few massive outliers.
* **StandardScaler** uses the mathematical *mean* and *variance*, both of which are heavily distorted by extreme values.
* **RobustScaler** scales data using the **Median** and the **Interquartile Range (IQR)** (the distance between the 25th and 75th percentiles). Because it ignores the extreme tails when calculating the scale, the core dataset remains perfectly distributed.

#### 2. It Centers Data Efficiently for Algorithms

Robust scaling transforms your continuous variables using this formula:

$$x_{\text{scaled}} = \frac{x - \text{median}}{\text{IQR}}$$

This centers the median of every feature exactly at $0$. Gradient descent-based algorithms (like Linear/Logistic Regression or Neural Networks) and distance-based algorithms (like KNN or SVM) converge significantly faster and remain mathematically stable when features share this uniform, zero-centered anchor point.

#### 3. It Preserves the Structural Variance of Real-World Data

Instead of clipping or artificially capping your highest-performing restaurants, Robust Scaling standardizes the scale while letting your outliers remain outliers. This allows your downstream machine learning models to still recognize that an ultra-expensive, highly-photographed restaurant is unique, without letting its massive raw scale corrupt the rest of the dataset's patterns.

Yes, a Zomato ratings dataset is **inherently and severely imbalanced.**

###Why the Dataset is Imbalanced
In online consumer platforms like Zomato, data naturally skews due to human behavior, creating a **highly asymmetrical distribution:**

*  **The "J-Shaped" or Right-Skewed Curve:** Most users only take the time to write reviews when they have a highly positive experience (4 or 5 stars) or an exceptionally terrible experience (1 star). Average or mediocre experiences (2 or 3 stars) are heavily underrepresented.

*  **Survival Bias:** Poorly rated restaurants lose customers quickly and go out of business, removing them from the active platform. High-rated restaurants survive, thrive, and continually accumulate a massive volume of reviews.

###Why Imbalance is Dangerous for Machine Learning

If you feed an imbalanced dataset straight into a classification model, the model will cheat.

If 85% of your data consists of 4-star ratings, an uncalibrated model can simply predict "4-stars" for every single restaurant. It will achieve an impressive **85% accuracy score** while being completely useless, as its ability to actually identify failing or poor restaurants is exactly 0%.

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

To handle the imbalanced dataset, the choice of technique depends entirely on whether the machine learning task is framed as a **Classification** task (predicting a categorical tier) or a **Regression** task (predicting an exact decimal rating).


### Technique 1: SMOTE (For Classification Tasks)

If predicting discrete categories like `Rating_Tier` (e.g., *Low, Medium, High*), we use **SMOTE (Synthetic Minority Over-sampling Technique)**.

* **Why:** Instead of just duplicating the rare minority classes (like 1-star reviews), which causes the model to overfit, SMOTE looks at the feature space feature-by-feature. It draws lines between existing minority examples and creates entirely new, synthetic data points along those lines. This forces the model to learn the structural decision boundaries of poorly rated restaurants rather than just ignoring them.


### Technique 2: Cost-Sensitive Learning / Sample Weights (For Regression Tasks)

If predicting the exact continuous `Rating` decimals (e.g., *3.7, 4.2*), SMOTE cannot be used. Instead, we use **Sample Weight Balancing**.

* **Why:** In regression, we cannot synthesize data points easily across continuous labels. Instead, we compute mathematical weights for each training row using `compute_sample_weight`. Rare ratings (like ultra-low 1.5 scores) are assigned a much higher weight, while overly common ratings (like 4.0 scores) get a lower weight. During training, if the model miscalculates a rare rating, it receives a heavy penalty in its loss function, forcing it to treat underrepresented ratings with equal importance.


### Summary of Benefit

Both techniques achieve the same core objective: they prevent the machine learning model from becoming heavily biased toward the dominant "highly-rated" restaurants, ensuring high predictive accuracy for both top-tier and struggling venues.

As outlined in the **Model Implementation** stage of your project architecture, we deployed **K-Means Clustering**.

K-Means is an unsupervised machine learning algorithm designed to discover hidden, natural groupings within data without relying on predefined labels. It works through an iterative optimization process:

1.  **Centroid Initialization:** The algorithm places a specified number of cluster centers ($k$) randomly or strategically (using k-means++) across the data space.
2.  **Distance Calculation & Assignment:** It calculates the geometric distance between every data point and all centroids, assigning each point to its closest center.
3.  **Centroid Update:** It shifts each centroid to the absolute mean coordinates of all data points assigned to it.
4.  **Convergence:** Steps 2 and 3 repeat sequentially until the centroids stop moving, minimizing the total **Inertia** (Within-Cluster Sum of Squares).

###2. Evaluating Performance: The Silhouette Score & Inertia
ChartBecause clustering works with unlabeled data, traditional metrics like Accuracy or $F_1$-Score cannot be used. Instead, we evaluate performance using **Inertia** (how tightly packed the clusters are) and the **Silhouette Score** (how well-separated the clusters are from each other).

The Silhouette Score ranges from -1 to +1, where a higher score indicates clean, distinct, and well-defined business segments.

### 1. Explanation of ML Model - 2: Agglomerative Hierarchical Clustering

Following the unsupervised segmentation roadmap in your project architecture, our second model is **Agglomerative Hierarchical Clustering**.

Unlike K-Means, which requires you to pre-specify the number of clusters up front, Agglomerative Clustering builds a bottom-up tree structure:

1. **Initial State:** It treats every single data point as an individual, isolated cluster of size one.
2. **Proximity Calculation:** It computes a distance matrix measuring the geometric proximity between all clusters using a specified metric (like Euclidean distance).
3. **Iterative Merging:** It searches for the two closest clusters and merges them into a single, larger cluster.
4. **Linkage Criterion:** It determines how distances between groups are measured (e.g., `Ward's linkage` minimizes the variance within the merging clusters). This step repeats sequentially until all data points are unified into a single root cluster.


### 2. Evaluating Performance: The Dendrogram Analysis

In your project architecture, the **Dendrogram** is the primary evaluation tool highlighted for hierarchical models. A dendrogram maps out every single merge configuration the algorithm made. The vertical length of the tree branches represents the distance between clusters; a longer vertical branch indicates that the merged groups were highly distinct from one another.



````python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

print("=== Phase 7.2: ML Model - 2 Implementation (Hierarchical Clustering) ===")

# ==========================================================
# 1. COMPUTE THE LINKAGE MATRIX FOR THE EVALUATION CHART
# ==========================================================
# 'X_scaled' is your preprocessed Min-Max data matrix
# 'ward' linkage minimizes the variance of clusters being merged
linkage_matrix = linkage(df_scaled[continuous_features], method='ward')

# ==========================================================
# 2. PLOT THE DENDROGRAM EVALUATION CHART
# ==========================================================
plt.figure(figsize=(10, 6))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',  # Shows only the final merged clusters for clean scannability
    p=12,                   # Truncate to show the last 12 merge branches
    leaf_rotation=45.,
    leaf_font_size=10.,
    show_contracted=True
)

plt.title("Hierarchical Clustering Dendrogram Evaluation Chart", fontsize=13, fontweight="bold")
plt.xlabel("Data Split Points / Cluster Index Size")
plt.ylabel("Distance Threshold (Cluster Dissimilarity)")
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()

# ==========================================================
# 3. FIT THE FINAL MODEL
# ==========================================================
# We choose 3 or 4 clusters based on where the longest vertical branches are cut
optimal_clusters = 3
hierarchical_model = AgglomerativeClustering(n_clusters=optimal_clusters, metric='euclidean', linkage='ward')
hierarchical_labels = hierarchical_model.fit_predict(df_scaled[continuous_features])

print(f" Hierarchical Clustering complete. Generated {optimal_clusters} distinct segments.")
````

### 1. Explanation of ML Model - 3: DBSCAN Clustering

The third model implemented in our marketplace segmentation pipeline is **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise).

Unlike K-Means and Hierarchical Clustering, which are distance-based and force every single data point into a cluster, DBSCAN relies on **spatial density**. It groups points together based on how tightly packed they are within a specific neighborhood radius, making it exceptionally powerful for discovering clusters of arbitrary shapes and naturally isolating anomalies.

The algorithm scans the dataset and classifies every point into one of three categories based on density rules:

1. **Core Points:** Points that contain at least a minimum number of neighbor points (`min_samples`) within a specified radius (`eps`).
2. **Border Points:** Points that fall inside the radius of a Core Point but do not contain enough neighbors of their own to form a new cluster core.
3. **Noise Points (Outliers):** Isolated points that do not fall within the radius of *any* Core Point. These are automatically assigned a cluster ID of `-1`.


### 2. Evaluating Performance: The Silhouette & Noise Trade-Off Chart

Because DBSCAN discards outliers as noise, evaluating its performance requires tracking how changing the radius (`eps`) impacts both the structural quality of the valid clusters (via the **Silhouette Score**) and the total percentage of data thrown out as **Marketplace Noise**.



````python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

print("=== Phase 8.3: Generating DBSCAN Operational Evaluation Charts ===")

# Range of epsilon radius values to test for structural density balance
eps_range = np.linspace(0.2, 0.6, 5)
silhouette_indices = []
noise_percentages = []

for e in eps_range:
    db = DBSCAN(eps=e, min_samples=6)
    labels = db.fit_predict(df_scaled[continuous_features])

    # Calculate noise percentage (Label == -1)
    noise_pct = (list(labels).count(-1) / len(labels)) * 100
    noise_percentages.append(noise_pct)

    # Calculate Silhouette score exclusively for clustered data (ignoring noise)
    if len(set(labels)) > 1 and (len(set(labels)) > 2 or -1 not in labels):
        valid_mask = labels != -1
        if np.sum(valid_mask) > 5:  # Ensure a baseline group remains
            score = silhouette_score(df_scaled[continuous_features][valid_mask], labels[valid_mask])
            silhouette_indices.append(score)
        else:
            silhouette_indices.append(0)
    else:
        silhouette_indices.append(0)
````

### **2. Cross- Validation & Hyperparameter Tuning**


For **DBSCAN**, traditional cross-validation cannot be applied directly because the algorithm is density-based and deterministic; it does not predict labels for unseen test sets using standard decision boundaries, nor does it take a fixed number of clusters as an input.

Instead, hyperparameter tuning focuses on evaluating a grid combination of **Epsilon** (eps) and **Minimum Samples** (min_samples). We optimize this grid using a custom scoring framework that targets the highest **Silhouette Coefficient** computed exclusively on the valid, non-noise clusters.

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

print("=== Phase 9.3: DBSCAN Density Hyperparameter Tuning Grid ===")

# ==========================================================
# 1. DEFINE HYPERPARAMETER SEARCH SPACE
# ==========================================================
# Testing a sweep of search radii (eps) and minimum density thresholds (min_samples)
param_grid_dbscan = {
    "eps": [0.25, 0.35, 0.45, 0.55],
    "min_samples": [4, 6, 8]
}

dbscan_tuning_results = []

# ==========================================================
# 2. ITERATING SEARCH SPACE FOR PARAMETER OPTIMIZATION
# ==========================================================
print(" Running Exhaustive Grid Optimization on density constraints...")

for e in param_grid_dbscan["eps"]:
    for ms in param_grid_dbscan["min_samples"]:

        # Initialize and fit the density model
        db = DBSCAN(eps=e, min_samples=ms)
        labels = db.fit_predict(df_scaled[continuous_features])

        # Count generated structure properties
        unique_labels = set(labels)
        n_clusters = len(unique_labels) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)
        noise_pct = (n_noise / len(labels)) * 100

        # Calculate silhouette quality index strictly for valid clusters (requires > 1 cluster)
        if n_clusters > 1:
            valid_mask = labels != -1
            # Ensure we have enough remaining data points to compute a valid silhouette score
            if np.sum(valid_mask) > ms:
                score = silhouette_score(df_scaled[continuous_features][valid_mask], labels[valid_mask])
            else:
                score = -1
        else:
            score = -1  # Invalid configuration if it collapses into 1 cluster or pure noise

        dbscan_tuning_results.append({
            "eps": e,
            "min_samples": ms,
            "n_clusters": n_clusters,
            "noise_percentage": noise_pct,
            "silhouette_score": score
        })

# ==========================================================
# 3. IDENTIFY AND EXTRACT BEST PARAMETERS
# ==========================================================
tuning_df_db = pd.DataFrame(dbscan_tuning_results)

# Filter out configurations that failed the cluster count requirement
valid_configs = tuning_df_db[tuning_df_db["silhouette_score"] > -1]

if not valid_configs.empty:
    best_db_config = valid_configs.sort_values(by="silhouette_score", ascending=False).iloc[0]

    print("\n Optimized Hyperparameters Found for Model 3:")
    print(f" -> Optimal Radius (eps)   : {best_db_config['eps']:.2f}")
    print(f" -> Min Core Samples       : {int(best_db_config['min_samples'])}")
    print(f" -> Natural Clusters Found : {int(best_db_config['n_clusters'])}")
    print(f" -> Marketplace Outlier %  : {best_db_config['noise_percentage']:.2f}%")
    print(f" -> Top Silhouette Score   : {best_db_config['silhouette_score']:.4f}")

    # ==========================================================
    # 4. DEPLOY CHAMPION DBSCAN MODEL
    # ==========================================================
    best_dbscan_model = DBSCAN(eps=best_db_config["eps"], min_samples=int(best_db_config["min_samples"]))
    dbscan_labels = best_dbscan_model.fit_predict(df_scaled[continuous_features])
    print("\n Optimized DBSCAN density map deployed successfully.")
else:
    print("\n[Warning]: No distinct multi-cluster configurations found in the grid. Adjust the parameter bounds.")
````

### **Which hyperparameter optimization technique have you used and why?**

For optimizing our **DBSCAN** density model, we utilized a custom **Exhaustive Grid Search Parameter Sweep** combined with a targeted **Noise-Filtering Silhouette Scorer**.


### Why Grid Optimization Was Used for DBSCAN

#### 1. Managing Interdependent Hyperparameters

DBSCAN’s behavior is strictly controlled by the structural relationship between two parameters: the search radius (`eps`) and the core density threshold (`min_samples`).

* **The Challenge:** These two parameters cannot be optimized independently. If you increase the search radius (`eps`), you must scale the neighborhood threshold (`min_samples`) accordingly to prevent all distinct groups from melting into a single giant cluster.
* **The Strategy:** A structured parameter grid search evaluates every pairing systematically. This ensures we pinpoint the exact threshold where clusters remain tightly packed without collapsing into a single macro-group.

#### 2. Isolating Anomalies via Custom Scoring

Standard unsupervised grid tuning calculates a silhouette score across an entire dataset indiscriminately.

* **The Problem:** Because DBSCAN assigns unclassifiable outliers a label of `-1` (noise), including those highly scattered points in a traditional validation calculation penalizes the model's score, making a good configuration look mathematically poor.
* **The Solution:** Our optimization framework applies a custom noise mask during the parameter sweep. It separates out the noise points and calculates the **Silhouette Coefficient** exclusively on the valid, dense clusters. This guarantees that our hyperparameter choice maximizes the clarity and definition of normal marketplace tiers while effectively isolating anomalies.

#### 3. Protecting Against Structural Collapse

If parameter boundaries are set blindly, density models can easily collapse, either grouping the entire marketplace into one cluster or classifying 100% of the merchants as outliers. By tracking `n_clusters` and `noise_percentage` side-by-side inside our tuning loop, the grid search filters out these invalid configurations automatically, locking in a balanced setup.


### Summary of Benefit

By implementing an exhaustive grid parameter sweep with outlier filtering, we isolated the precise density rules needed to group regular marketplace partners into distinct, compact tiers while simultaneously establishing a reliable, automated pipeline to catch operational anomalies.
